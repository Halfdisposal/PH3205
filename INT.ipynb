```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def forward_diff(f, xn, h):
    return (f(xn + h) - f(xn)) / h

def backward_diff(f, xn, h):
    return (f(xn) - f(xn - h)) / h

def central_diff(f, xn, h):
    return (f(xn + h) - f(xn - h)) / (2 * h)

def five_point_approx(f, xn, h):
    return (f(xn - 2 * h) - 8 * f(xn - h) + 8 * f(xn + h) - f(xn + 2 * h)) / (12 * h)
```

```python
def Trapezoid(f, x_min, x_max, N):
    h = (x_max - x_min)/(N - 1)
    x, h = np.linspace(x_min, x_max, N, retstep=True)
    fx = f(x)
    return 0.5*h*(fx[0] + fx[-1] + 2.0*np.sum(fx[1:-1])), h

def Simpson13(f, x_min, x_max, N):
    if N%2 == 0:
        print("N has to be odd")
        return None
    h = (x_max - x_min)/(N - 1)
    x, h = np.linspace(x_min, x_max, N, retstep=True)
    fx = f(x)
    #print(x, '\n', x[2:-1:2], '\n', x[1:-1:2])
    return h*(fx[0] + fx[-1] + 4*np.sum(fx[1:-1:2]) + 2*np.sum(fx[2:-1:2]))/3.0, h

def Simpson38(f, x_min, x_max, N):
    h = (x_max - x_min)/(N - 1)
    x, h = np.linspace(x_min, x_max, N, retstep=True)
    fx = f(x)
    return 3*h*(fx[0] + fx[-1] + 3.0*np.sum(fx[1:-1:3]) + 3.0*np.sum(fx[2:-1:3]) + 2.0*np.sum(fx[3:-1:3]))/8.0, h

def Boole(f, x_min, x_max, N):
    h = (x_max - x_min)/(N - 1)
    x, h = np.linspace(x_min, x_max, N, retstep=True)
    fx = f(x)
    return 2.0*h*(7.0*fx[0] + 7.0*fx[-1] + 32.0*np.sum(fx[1:-1:2]) + 12.0*np.sum(fx[2:-1:4]) + 14*np.sum(fx[4:-1:4]))/45.0, h
```

```python
def f(x):
    return np.sin(x)

xmin = 0.0
xmax = np.pi / 2.0
N = 1001
i_exact = -(np.cos(xmax) - np.cos(xmin))
i_trap = Trapezoid(f, xmin, xmax, N)[0]
i_simp13 = Simpson13(f, xmin, xmax, N)[0]
i_simp38 = Simpson38(f, xmin, xmax, N)[0]

print(f"{i_exact}\t{i_trap}\t{i_simp13}\t{i_simp38}")
```
